
## Problem 6-1 a

```python
a = np.array([1, 1, 1])
b = np.array([-1, 1, 1])
c = np.array([1, -1, 1])
```

```python
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)
norm_c = np.linalg.norm(c)

print(norm_a, norm_b, norm_c)
```

```python
dot_ab = np.dot(a, b)
dot_ac = np.dot(a, c)
dot_bc = np.dot(b, c)

print(dot_ab, dot_ac, dot_bc)
```

```python
def unit_vector(v):
    return v / np.linalg.norm(v)

def angle_between(v1, v2):
    v1_u = unit_vector(v1)
    v2_u = unit_vector(v2)

    return np.arccos(np.clip(np.dot(v1_u, v2_u), -1, 1))
```

```python
theta_ab = angle_between(a, b) 
theta_ac = angle_between(a, c)
theta_bc = angle_between(b, c)

print(theta_ab, theta_ac, theta_bc)
```
